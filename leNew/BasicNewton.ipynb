```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import myModule
from myModule.interactions import newton
```

```python
mySim = myModule.simulation('odeint')
newton1 = myModule.interaction('Newton in 1', newton.acceleration, center = np.array([1,0,0]))
newton2 = myModule.interaction('Newton in 0', newton.acceleration, center = np.array([0,0,0]))
mySim.interactions.append(newton1)
mySim.interactions.append(newton2)
mySim.updateAcceleration()
```

```python
s_0 = np.array([1, 1, 0, 1, 0 ,0])
t_i = 0
t_f = 200
dt = 0.01
mySim.integrate(s_0, t_i, t_f, dt)
```
