```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sim
```

```python
def f(s,t):
    ans = np.empty(6) 
    ans[3:] = [1,0,0]-s[:3]
    ans[:3] = 0
    ans[3:] /= np.sum(ans[3:]**2)**(3/2)
    return ans

def g(s,t):
    ans = np.empty(6) 
    ans[3:] = [-1,0,0]-s[:3]
    ans[:3] = 0
    ans[3:] /= np.sum(ans[3:]**2)**(3/2)
    return ans
```

```python
simulacion = sim.Simulation('odeint')
simulacion.addForce('NewtonIn+1','Conservative', f)
simulacion.addForce('NewtonIn-1','Conservative', g)
```

```python
s0 = np.array([1,1,0,1,0,0])
```

```python
simulacion.integrate(s0, 0, 200, 0.01)
```

```python
plt.plot(simulacion.path[:,0],simulacion.path[:,1])
plt.plot([1,-1],[0,0],'ok')
```
